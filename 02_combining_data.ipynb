# 02 Combining the Dataframes

```python
import pandas as pd
```

## 2.1 Nat. Disasters - Importing Data

```python
nat_df = pd.read_csv('data/all_natural_disasters_reduced.csv', index_col=0)
nat_df.head()
```

## 2.2 GDP's - Importing Data

```python
gdps_df = pd.read_csv('data/country_gdps.csv', index_col='iso').drop('Unnamed: 0', axis=1)
gdps_df.head()
```

## 2.3 Combining Datasets
In this step the natural disasters dataframes will be extended which 5 rows which hold the gpd's of the previous, current as well as the 3 following years.

- GDP-1, GDP, GDP+1, GDP+2, GDP+3
- Since we have GDP Data from 1960 to 2021 we can use natural disasters from 1961 to 2018

```python
df_combined = nat_df.copy()

for index, row in df_combined.iterrows():
    iso = row['ISO']
    year = row['Year']

    df_combined.at[index, 'GDP-1'] = gdps_df.at[iso, f'{year-1}']
    df_combined.at[index, 'GDP'] = gdps_df.at[iso, f'{year}']
    df_combined.at[index, 'GDP+1'] = gdps_df.at[iso, f'{year+1}']
    df_combined.at[index, 'GDP+2'] = gdps_df.at[iso, f'{year+2}']
    df_combined.at[index, 'GDP+3'] = gdps_df.at[iso, f'{year+3}']

df_combined.head()
```

## 2.4 Cleaning Up
Dropping Rows with Null Values in Target Columns

```python
df_combined = df_combined.dropna(subset=['GDP-1', 'GDP', 'GDP+1', 'GDP+2', 'GDP+3'])
df_combined.info()
```

## 2.5 Save Dataset

```python
df_combined.to_csv('data/nat_disasters_vs_gdp.csv')
```
